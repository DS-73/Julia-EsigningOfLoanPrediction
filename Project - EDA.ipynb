```julia
y = df[:e_signed]

df2 = df
select!(df2, Not([:entry_id, :pay_schedule, :e_signed]));
elements = names(df2)
```

# Plots

```julia
plotd = []
for i in elements
    push!(plotd, histogram(
        df2[i],
        marker = (0, 0, [:none :+ :none]),
        markerstrokewidth = 0,
        color = [:steelblue :orangered :green],
        fill = 0.5,
        orientation = [:v :v :h],
        title = i,
        legend = false
    ))
end
```

# Histogram

```julia
vector1 = [0]
vector0 = [1]

for val in y
    if(val == 1)
        push!(vector1, 1)
    else
        push!(vector0, 0)
    end
end
```

```julia
histogram(vector0, label="0", bins= 3, title = "Label Count Histogram")
histogram!(vector1, label="1", bins= 4, color = "orange")
```
